```python
df3=sns.load_dataset("dots")
df3
```

```python
sns.violinplot(x="align",y="firing_rate",data=df3)
```

```python
sns.violinplot(x="align",y="firing_rate",data=df3,hue="choice")
```

```python
sns.violinplot(x="align",y="firing_rate",data=df3,hue="choice")
sns.violinplot(x="align",y="firing_rate",data=df3,hue="choice",dodge=False)
```

## KDE Plot

```python
sns.kdeplot(x="firing_rate",data=df3)
```

```python
sns.jointplot(x="firing_rate",y="coherence",data=df3,hue="time")
```

```python
sns.jointplot(x="firing_rate",y="coherence",data=df3,kind="reg")
```

```python
sns.jointplot(x="firing_rate",y="coherence",data=df3,kind="hex")
```
