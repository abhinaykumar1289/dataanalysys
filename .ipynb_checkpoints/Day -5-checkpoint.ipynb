```python
df1=pd.DataFrame({"city":["cpt","nrt","gnt","vij"],"temperature":[33,36,28,40]})
df1
```

```python
df2=pd.DataFrame({"city":["cpt","nrt","gnt","vij"],"humidity":[20,25,30,32]})
df2
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
df1=pd.DataFrame({"city":["cpt","nrt","gnt","vij"],"temperature":[33,36,28,40]})
df1
```

```python
df2=pd.DataFrame({"city":["ong","hyd","gnt","vij"],"humidity":[20,25,30,32]})
df2
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right",indicator=True)
df3
```
